# Cuckoo search()
How does it work:

> $x_{t-1} = x_t + sE_t$
>
> Where $E_t$ is drawn from a standard normal distribution with zero mean and unity standard deviation for random walks
>
>[wiki](https://en.wikipedia.org/wiki/Cuckoo_search)

```python
from mealpy.swarm_based.CSA import OriginalCSA

def rosenbrock_function(position):
    value = [
        100 * (position[i+1] - position[i]**2)**2 +
        (1-position[i])**2
        for i in range(len(position)-1)
    ]
    return sum(value)

problem_dict1 = {
    "fit_func": rosenbrock_function,
    "lb": [-10, -10, -5],
    "ub": [10, 10, 5],
    "minmax": "min",
    "log_to": "log.txt",
}

epoch = 1000
pop_size = 100
model = OriginalCSA(epoch=epoch, pop_size=pop_size)
best_position, best_fitness = model.solve(problem_dict1)
print(f"Solution: {best_position}, Fitness: {best_fitness}")
```
